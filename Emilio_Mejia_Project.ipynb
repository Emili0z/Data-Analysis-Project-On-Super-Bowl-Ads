```python
#Dropping columns that are not necessary to data.
sb = sb.drop(['Superbowl Ads Link','Danger','Uses Sex','Length','Estimated Cost','Youtube Likes'], axis = 1)
```

```python
#Display newlook data set
sb.head(5)
```

```python
#Sorting by YouTube views to display return of viewership by audiences.
sb =sb.sort_values(by=['Youtube Views'], ascending=False)
sb
```

```python
#Changing boolean T/F values to integers for plot purposes.
sb['Funny'] = sb['Funny'].astype(int)
sb['Shows Product Quickly'] = sb['Shows Product Quickly'].astype(int)
sb['Patriotic'] = sb['Patriotic'].astype(int)
sb['Celebrity'] = sb['Celebrity'].astype(int)
sb['Animals'] = sb['Animals'].astype(int)
sb
```

```python
#Creating a bar chart that displays the correction of how a celebrity being in a Super Bowl commercial affects whether viewers return to watch the advertisement later on YouTube.

plt.figure(figsize=(10,5))
sns.barplot(data = sb,
            x = 'Celebrity',
            y = 'Youtube Views',
            hue = 'Celebrity',
            color = 'b')
plt.title("Celebrity Influence on Returning to Super Bowl Commercials")
```

```python
#Boxplots for how Patriotic commercials and Animals in the commercials affect the return of viewers to the Super Bowl ads.
plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
sns.boxplot(data = sb, x = 'Patriotic', y = 'Youtube Views', hue = 'Patriotic')
plt.title("Super Bowl Patriotic Commercials vs. Return Viewing")

plt.subplot(1,3,2)
sns.boxplot(data = sb, x = 'Animals', y = 'Youtube Views', hue = 'Animals')
plt.title("Super Bowl Animal Commercials vs. Return Viewing")
```

```python
#Strip plot to represent the relationship between YouTube views and if the commercial shows the product quickly
sns.stripplot(data = sb, x = 'Shows Product Quickly', y = 'Youtube Views', hue = "Shows Product Quickly", palette = "deep")
```

```python
#Showing a jointplot to see whether the relationship is negative or positive between
plot = sns.jointplot(x = 'Funny', y = 'Youtube Views', data = sb, hue = 'Funny')
plot.fig.suptitle('Comedy Affecting Super Bowl Return Viewership')
```

```python
#The joint plot looks less normal than usual joint plots.
#This is due to the disparity of high numbers vs. the low 0, 1 for true or false for 'Funny'
#At the tip, the Funny presents a positive relationship as the highest viewer
#count is represented through a funny advertisement.
```
